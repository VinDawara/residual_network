```python
read_path = r'C:\MyData\Stressed network work\Data\simulation\dissipative_energy'
save_path = r'C:\MyData\Stressed network work\Data\simulation\dissipative_energy\plots'

with open(read_path + '/T3_m5eb03h5_d5t125.txt', 'r') as f:
    data = f.readlines()
time2 = np.zeros(len(data))
energy2 = np.zeros(len(data))
cracklength2 = np.zeros(len(data))
for i, line in enumerate(data):
    time2[i], cracklength2[i], energy2[i] = line.split()

with open(read_path + '/T4_m5eb035h5_d5t125.txt', 'r') as f:
    data = f.readlines()

time3 = np.zeros(len(data))
energy3 = np.zeros(len(data))
cracklength3 = np.zeros(len(data))
for i, line in enumerate(data):
    time3[i], cracklength3[i], energy3[i] = line.split()

with open(read_path + '/T5_m5eb04h5_d5t125.txt', 'r') as f:
    data = f.readlines()

time5 = np.zeros(len(data))
energy5 = np.zeros(len(data))
cracklength5 = np.zeros(len(data))
for i, line in enumerate(data):
    time5[i], cracklength5[i], energy5[i] = line.split()

L = 109.5
fig, ax = plt.subplots()
ax.plot(time2/L, energy2/energy2[0], color = 'tab:blue', label=r'$\epsilon_{b} = 0.030$', 
        alpha=1, linestyle = 'None', marker='o', markersize=8, markeredgecolor='black')
ax.plot(time3/L, energy3/energy3[0], color = 'tab:orange', label=r'$\epsilon_{b} = 0.035$', 
        alpha=1, linestyle = 'None', marker='o', markersize=8, markeredgecolor='black')
ax.plot(time5/L, energy5/energy5[0], color = 'tab:green', label=r'$\epsilon_{b} = 0.040$', 
        alpha=1, linestyle = 'None', marker='o', markersize=8, markeredgecolor='black')
ax.set_xlabel(r'$2t c_r/L \rightarrow $')
ax.set_ylabel(r'$U(t)/U_0\rightarrow$')
# ax.legend()
ax.set_xlim(left = 0)
ax.set_ylim(bottom = 0)
fig.savefig(save_path+ '/total_energy_vary_eb.png', bbox_inches = 'tight', dpi = 300)
plt.show()
```
